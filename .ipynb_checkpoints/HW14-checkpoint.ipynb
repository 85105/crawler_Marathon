## 爬取所有上映中的電影資料：電影名稱、期待度、滿意度、簡介等等

```python
import requests
from bs4 import BeautifulSoup
from fake_useragent import UserAgent
url = "https://movies.yahoo.com.tw/movie_intheaters.html"
UA = UserAgent()
header = {'user-agent': UA.random}
response = requests.get(url , headers = header)
soup = BeautifulSoup(response.text,'lxml')

infor = soup.find('ul' , attrs={'class':'release_list'}).find_all('li')

for i in infor:
    name = i.find('div', attrs={'class':'release_movie_name'})
    level = i.find('dt').find('div', attrs={'class':'leveltext'})
    level_sati = i.find('dd').find('div', attrs={'class':'leveltext'})
    text = i.find('div', attrs={'class':'release_text'})
    print(name.a.text.strip())
    print('期待度: ',level.span.text.strip())
    print('滿意度: ',level_sati.span["data-num"].strip())
    print(text.span.string.strip(),'\n')
```

```python
# 先點選網頁最底部的下一頁，觀察其網址變化。然後我們把頁數當作變數傳入，用迴圈爬取。

import requests
from bs4 import BeautifulSoup
from fake_useragent import UserAgent
max_page = 9
for page_number in range(1, max_page+1, 1):
    url = "https://movies.yahoo.com.tw/movie_intheaters.html"
    UA = UserAgent()
    payload = {'page':str(page_number)}
    header = {'user-agent': UA.random}
    response = requests.get(url , headers = header, params=payload)
    soup = BeautifulSoup(response.text,'lxml')

    infor = soup.find('ul' , attrs={'class':'release_list'}).find_all('li')

    for i in infor:
        name = i.find('div', attrs={'class':'release_movie_name'})
        level = i.find('dt').find('div', attrs={'class':'leveltext'})
        level_sati = i.find('dd').find('div', attrs={'class':'leveltext'})
        text = i.find('div', attrs={'class':'release_text'})
        print(name.a.text.strip())
        print('期待度: ',level.span.text.strip())
        print('滿意度: ',level_sati.span["data-num"].strip())
        print(text.span.string.strip(),'\n')
```
