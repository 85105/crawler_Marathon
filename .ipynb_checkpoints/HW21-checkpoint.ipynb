# 打開瀏覽器

```python
from selenium import webdriver
from bs4 import BeautifulSoup
browser = webdriver.Chrome(executable_path='chromedriver')
browser.get("https://www.ettoday.net/news/news-list.htm")
```

## 這邊的 10000 是故意設一個很大的數字，讓程式一直想拉到最底的概念

```python
import time
for i in range(10):
    time.sleep(1)
    browser.execute_script("window.scrollTo(0, 10000);")
```

## browser.page_source 取得網頁上的資料

```python
html_source = browser.page_source
soup = BeautifulSoup(html_source, "html5lib")

for d in soup.find(class_="part_list_2").find_all('h3'):
    print(d.find(class_="date").text, d.find_all('a')[-1].text)
```

## 關閉瀏覽器

```python
browser.quit();
```

# 動態抓網頁，全部code

```python
from selenium import webdriver
from bs4 import BeautifulSoup
import time
browser = webdriver.Chrome(executable_path='chromedriver')
browser.get("https://www.ettoday.net/news/news-list.htm")


for i in range(10):
    time.sleep(1)
    browser.execute_script("window.scrollTo(0, 10000);")
    
html_source = browser.page_source
soup = BeautifulSoup(html_source, "html5lib")

for d in soup.find(class_="part_list_2").find_all('h3'):
    print(d.find(class_="date").text, d.find_all('a')[-1].text)
browser.quit();
```

## 取出今天所有的新聞

```python
from selenium import webdriver
from bs4 import BeautifulSoup
import time
import datetime

loc_dt = datetime.datetime.today() 
loc_dt_format = loc_dt.strftime("%Y/%m/%d")


browser = webdriver.Chrome(executable_path='chromedriver')
browser.get("https://www.ettoday.net/news/news-list.htm")


for i in range(10):
    time.sleep(2)
    browser.execute_script("window.scrollTo(0, 10000);")
    
html_source = browser.page_source
soup = BeautifulSoup(html_source, "html5lib")

for d in soup.find(class_="part_list_2").find_all('h3'):
    # d.find(class_="date").text   => 是字串，但包含日期及時間
    t = d.find(class_="date").text      
    # 篩掉時間的部分
    if(t[:10] == loc_dt_format):
        print(d.find(class_="date").text, d.find_all('a')[-1].text)
    
    
browser.quit();
```

## 取出現在時間兩小時內的新聞

```python
from selenium import webdriver
from bs4 import BeautifulSoup
import time
import datetime

loc_dt = datetime.datetime.today() 
time_del = datetime.timedelta(hours=-2)
new_dt = loc_dt + time_del 
datetime_format = new_dt.strftime("%Y/%m/%d %H:%M")
loc_dt_format = loc_dt.strftime("%Y/%m/%d")


browser = webdriver.Chrome(executable_path='chromedriver')
browser.get("https://www.ettoday.net/news/news-list.htm")


for i in range(10):
    time.sleep(2)
    browser.execute_script("window.scrollTo(0, 10000);")
    
html_source = browser.page_source
soup = BeautifulSoup(html_source, "html5lib")

for d in soup.find(class_="part_list_2").find_all('h3'):
    # d.find(class_="date").text   => 是字串，但包含日期及時間
    t = d.find(class_="date").text  
    # 篩掉時間的部分
    if(t[:10] == loc_dt_format and t[11:13]>=datetime_format[11:13]):
        print(d.find(class_="date").text, d.find_all('a')[-1].text)
    
    
browser.quit();
```

## 根據範例，取出三天前下午三點到五點的新聞 

```python
from selenium import webdriver
from bs4 import BeautifulSoup
import time
import datetime

loc_dt = datetime.datetime.today() 
time_del = datetime.timedelta(days=-3)
new_dt = loc_dt + time_del 
datetime_format = new_dt.strftime("%Y-%m-%d")
loc_dt_format = new_dt.strftime("%Y/%m/%d")


browser = webdriver.Chrome(executable_path='chromedriver')

info = 'https://www.ettoday.net/news/news-list-'+datetime_format+'-0.htm'

browser.get(info)


for i in range(10):
    time.sleep(1)
    browser.execute_script("window.scrollTo(0, 10000);")
    
html_source = browser.page_source
soup = BeautifulSoup(html_source, "html5lib")

for d in soup.find(class_="part_list_2").find_all('h3'):
    # d.find(class_="date").text   => 是字串，但包含日期及時間
    t = d.find(class_="date").text  
    # 篩掉時間的部分
    if(t[:10] == loc_dt_format and '15' <= t[11:13] <='17'):
        print(d.find(class_="date").text, d.find_all('a')[-1].text)
    
    
browser.quit();
```
